```python
import pandas as pd
df=pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
print("Rows:",df.shape[0])
print("columns:",df.shape[1])
```

```python
df['Online Store']=df['Online Store'].fillna('In-store Purchase')
```

```python
df['Size']=df['Size'].fillna('Free Size') 
df.iloc[30:41].head(10)
```

```python
df['Delivery Speed']=df['Delivery Speed'].fillna('N/A(Offline)')
```

```python
df.isnull().sum()
```

```python
labels=['Young Adult','Adult','Middle-aged','senior']
df['Age group']=pd.qcut(df['Age'],q=4,labels=labels)
```

```python
df[['Age','Age group']].head(10)
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
